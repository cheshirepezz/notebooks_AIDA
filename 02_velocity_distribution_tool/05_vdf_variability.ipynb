# Variability of the electron VDF in the magnetosheath - an eventful case
___
![illustration](graphics/illustration_variability.png)
___
This notebook illustrates how in less than a minute worth of MMS data the electron VDF can profoundly transform multiple time. We show 6 of these very characteristic distributions, and also show how the scaled and normalised views complete the picture, by highlighting these VDF features. We do not discuss the relation of these distributions to waves, nore their stability, and leave these as open questions.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import sys, os
import warnings
warnings.filterwarnings('ignore')
```

```python
from aidapy import load_data
```

```python
start_time = datetime(2015, 9, 19, 7, 40, 24)
end_time   = datetime(2015, 9, 19, 7, 44, 0)
settings = {'prod': ['i_dist', 'e_dist', 'dc_mag', 'sc_att', 'i_bulkv'],
            'probes': ['3'], 'coords': 'gse', 'mode': 'high_res', 'frame':'gse'}
```

```python
xr_mms = load_data(mission='mms', start_time=start_time, end_time=end_time, **settings)
```

```python
species      = 'electron'
frame        = 'B'
grid_geom     = 'spher'
v_max        = 1.6e7
resolution   = 120
interp_schem = 'cub'
```

```python
start_time_sub = datetime(start_time.year, start_time.month, start_time.day, 7, 43, 49, 699000)
end_time_sub   = datetime(start_time.year, start_time.month, start_time.day, 7, 43, 49, 819000)
xr_mms = xr_mms.vdf.interpolate(start_time, end_time, start_time_sub, end_time_sub,
                                species=species, frame=frame, grid_geom=grid_geom,
                                v_max=v_max, resolution=resolution, interp_schem=interp_schem, verbose=False)
xr_mms.vdf.plot('1d')
xr_mms.vdf.plot('2d', plt_contourf=True)
```

```python
start_time_sub = datetime(start_time.year, start_time.month, start_time.day, 7, 43, 50, 59000)
end_time_sub   = datetime(start_time.year, start_time.month, start_time.day, 7, 43, 50, 359000)
xr_mms = xr_mms.vdf.interpolate(start_time, end_time, start_time_sub, end_time_sub,
                                species=species, frame=frame, grid_geom=grid_geom,
                                v_max=v_max, resolution=resolution, interp_schem=interp_schem, verbose=False)
xr_mms.vdf.plot('1d')
xr_mms.vdf.plot('2d', plt_contourf=True)
```
